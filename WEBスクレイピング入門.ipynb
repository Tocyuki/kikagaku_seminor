## æ˜Ÿã®æ•°ã‚’å–å¾—

```python
stars = []
for i in range(len(rankings)):
    stars.append(float(rankings[i].find_element_by_class_name('evaluateNumber').text))
```

## è¦³å…‰åœ°åã¨è©•ä¾¡ã‚’ç´ä»˜ã‘ã‚‹è¾æ›¸ã‚’ä½œæˆ
