#### 実践課題
下記のcandidatesとfor文を使って、各アンケート結果に対する回答数をhist_dictの辞書に保存して下さい。

```python
df = pd.read_csv('data2.csv', skiprows=1, header=None)
candidates = [
    '感じている',
    'どちらかと言えば感じている',
    'どちらかと言えば感じていない',
    '感じていない'
]
```

```python
hist_dict = {
    candidates[0]: 0,
    candidates[1]: 0,
    candidates[2]: 0,
    candidates[3]: 0
}

for df in df[29]:
    for candidate in candidates:
        if df == candidate:
            hist_dict[candidate] += 1
```

```python
hist_dict
```

```python
# 回答
hist_dict = {}
for candidate in candidates:
    count = (df[29] == candidate).sum()
    hist_dict[candidate] = count
```

# 全職業でプロットする

```python
# 関数化
def plot_pie(dataframe):
    # candidatesの定義
    candidates = ['感じている', 'どちらかと言えば感じている', 'どちらかと言えば感じていない', '感じていない']
    # dfを処理
    dict_hist = {}
    for candidate in candidates:
        count = (dataframe['仕事が機械に奪われるという危機感を感じるか'] == candidate).sum()
        dict_hist[candidate] = count
    
    # 数値を抽出
    hist = []
    for val in dict_hist.values():
        hist.append(val)
    
    # 0があれば除外
    plot_dict_hist = {}
    for key, value in dict_hist.items():
        if value != 0:
            plot_dict_hist[key] = value
            
    # 円グラフをプロット
    plt.pie(plot_dict_hist.values(), autopct='%.f%%', startangle=90, counterclock=False, labels=plot_dict_hist.keys())
    plt.axis('equal') 
    plt.show()
```

```python
# 全職種のリストを作成
work_kinds = df1['職業'].unique()
```

```python
for work_kind in work_kinds:
    dataframe = df1 [ df1['職業'].isin([work_kind])]
    print(work_kind)
    plot_pie(dataframe)
```
